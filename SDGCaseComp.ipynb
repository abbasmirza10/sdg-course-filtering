```python
import pandas as pd
import numpy as np
```

```python
def filter_dataframe(csv_url, keywords):
    df = pd.read_csv(csv_url)
    df = df.dropna(subset=['Section Info'])
    df = df.drop_duplicates(subset=['Name'])
    df = df[~df['Section Info'].str.startswith('Same as')]
    filtered_df = pd.DataFrame()
    for keyword in keywords:
        regex_pattern = r'\b{}\b'.format(keyword)
        filtered_rows = df[df['Description'].str.contains(regex_pattern, case=False, na=False, regex=True)]
        filtered_df = pd.concat([filtered_df, filtered_rows])
    return filtered_df
```
